```python
import os
from glob import glob

# Write your code below.

# Load the environment variable
price_data_path = os.getenv("PRICE_DATA")
print("PRICE_DATA path:", price_data_path)

# Check it's not None
assert price_data_path is not None, "PRICE_DATA environment variable is not set."

# Use glob to find all parquet files in the directory
parquet_files = glob(os.path.join(price_data_path, "*.parquet"))


# Recursively search for .parquet files in all subdirectories
parquet_files = glob(os.path.join(price_data_path, "**", "*.parquet"), recursive=True)



# Optional: Print the list of files
print("Parquet files found:", parquet_files)
```

For each ticker and using Dask, do the following:

+ Add lags for variables Close and Adj_Close.
+ Add returns based on Close:
    
    - `returns`: (Close / Close_lag_1) - 1

+ Add the following range: 

    - `hi_lo_range`: this is the day's High minus Low.

+ Assign the result to `dd_feat`.

(4 pt)

```python


# Load the parquet files
ddf = dd.read_parquet(parquet_files)

# Ensure dates are in datetime format
ddf['Date'] = dd.to_datetime(ddf['Date'])

# Sort the data for lag calculations
ddf = ddf.sort_values(by=["ticker", "Date"])

# Group by ticker and apply feature engineering
dd_feat = ddf.groupby("ticker").apply(
    lambda df: df.assign(
        Close_lag_1=df["Close"].shift(1),
        Adj_Close_lag_1=df["Adj Close"].shift(1),
        returns=(df["Close"] / df["Close"].shift(1)) - 1,
        hi_lo_range=df["High"] - df["Low"]
    ),
    meta={
        'Date': 'datetime64[ns]',
        'Open': 'float64',
        'High': 'float64',
        'Low': 'float64',
        'Close': 'float64',
        'Adj Close': 'float64',
        'Volume': 'float64',
        'source': 'object',
        'ticker': 'object',
        'Year': 'int64',
        'Close_lag_1': 'float64',
        'Adj_Close_lag_1': 'float64',
        'returns': 'float64',
        'hi_lo_range': 'float64'
    }
)
```

+ Convert the Dask data frame to a pandas data frame. 
+ Add a new feature containing the moving average of `returns` using a window of 10 days. There are several ways to solve this task, a simple one uses `.rolling(10).mean()`.

(3 pt)

```python
ddf = dd.read_parquet(parquet_files)
print(ddf.columns)
```

```python
# Convert Dask to pandas
pdf = dd_feat.compute()

# Fix: Reset index to remove multi-index ambiguity
pdf = pdf.reset_index(drop = True)

# Sort values properly
pdf = pdf.sort_values(by=["ticker", "Date"])

# Calculate 10-day moving average of returns
pdf["returns_ma_10"] = pdf.groupby("ticker")["returns"].transform(lambda x: x.rolling(10).mean())

pdf.head()
```

+ Was it necessary to convert to pandas to calculate the moving average return?
+ Would it have been better to do it in Dask? Why?

(1 pt)

>>> ### It wasn’t strictly necessary to convert to pandas, but it made calculating the moving average easier due to pandas’ mature groupby().rolling() support. 

>>> ### Doing it in Dask would be better for large datasets, but it requires more setup, like proper partitioning, to work correctly. Dask is designed for scalability. If your data doesn’t fit into memory, keeping everything in Dask allows distributed computation and prevents memory overload. However, Dask’s rolling with groupby is less mature and trickier to use than pandas.
